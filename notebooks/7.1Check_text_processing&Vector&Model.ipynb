```python
import numpy as np
import matplotlib.pyplot as plt

# Define the metrics for Random Forest and Support Vector Machine
metrics = {
    'Random Forest': {
        'precision': [0.96, 0.96],
        'recall': [0.96, 0.96],
        'f1-score': [0.96, 0.96],
        'support': [4650, 4330]
    },
    'Support Vector Machine': {
        'precision': [0.97, 0.95],
        'recall': [0.96, 0.97],
        'f1-score': [0.96, 0.96],
        'support': [4650, 4330]
    }
}

# Categories and bar width
categories = ['Fake', 'True']
bar_width = 0.35

# Create subplots
fig, axes = plt.subplots(1, 4, figsize=(20, 6), sharey=False)

# Plot each metric
for i, (metric, ax) in enumerate(zip(['precision', 'recall', 'f1-score', 'support'], axes)):
    rf_values = metrics['Random Forest'][metric]
    svm_values = metrics['Support Vector Machine'][metric]

    # Positions of the bars on the x-axis
    r1 = np.arange(len(rf_values))
    r2 = [x + bar_width for x in r1]

    # Create bar plots
    ax.bar(r1, rf_values, color='lightgreen', width=bar_width, edgecolor='grey', label='Random Forest')
    ax.bar(r2, svm_values, color='lightsalmon', width=bar_width, edgecolor='grey', label='Support Vector Machine')

    # Add titles and labels
    ax.set_title(metric.capitalize())
    ax.set_xticks([r + bar_width / 2 for r in range(len(rf_values))])
    ax.set_xticklabels(categories)
    if metric != 'support':
        ax.set_ylim(0, 1)

    # Adding the legend
    if i == 0:
        ax.legend()

# Display the plot
plt.tight_layout()
plt.show()
```
